## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
data.boxplot(column='price', by='grade')

plt.title("Boxplot of Apartment Price by Grade")
plt.suptitle("")  # Usuwamy domyÅ›lny tytuÅ‚
plt.xlabel("Grade")
plt.ylabel("Price")

plt.ticklabel_format(style='plain', axis='y')
#Hint: plt.boxplot
```

Boxplot can be very complex so lets have a look at each element.

The line within each box represents the median price for specific grade.

The box itself show the interquartile range, indicating where the middle 50% of prices fall for each grade. Wider boxes suggest a larger spread of prices within that grade.

Whiskers extend from the box, showing the range of prices excluding potential outliers. Longer whiskers indicate a wider range of prices.

The individual points beyond the whiskers represent potential outliers, which are apartaments with ususally high or low prices for their grade.

Comparing this plot with plot 2b, we can still notice the trend of price increase with the increase in grade, however, we are provided with more information.


```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:

plt.figure(figsize=(8, 6))
mosaic(data, ['bedrooms', 'grade'], title='Mosaic Plot of Bedrooms vs. Grade')
plt.show()

```

The Mosaic Plot works in such a way that the larger the area of a particular rectangle, the more occurrences we have that meet the given condition. From the above chart, we can see that we have the most offers with three bedrooms and a grade of 7. At the edges, the chart becomes illegible due to the small number of offers.

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
interested_in = data.loc[(data['bedrooms'] == 3) & (data['bathrooms'] == 2) & (data['floors'] == 2)]

interested_in.describe()
#Hint: pandas.DataFrame.loc
```

We have 154 offers meeting our expectations. On average, we would have to pay around 514498 for the apartment we are interested in. The maximum square foot living that we can find with the criteria we are interested in is 3650.

```python
#b)
#Your code:
plt.scatter(x=data['grade'], y=data['price'], color='black', alpha=0.5, s=0.75)
plt.scatter(x=interested_in['grade'], y=interested_in['price'], color='yellow', alpha=0.5)
plt.ticklabel_format(style='plain')
```

Offers meeting our criteria are visualized in yellow on the price by grade chart. We can notice that the apartments we are interested in are in grades 6 to 10.

```python
#c)
#Your code:
interested_in.loc[(interested_in['price'] / interested_in['sqft_living']).sort_values().index].head()
#Hint: .loc, .sort_values(), .index
```

The above data is sorted in ascending order. Thanks to this, we will get the smallest price to area ratio as the first result. The smallest ratio is the best because we assume that we are a customer interested in buying such a house.
